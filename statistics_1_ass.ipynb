# Question 7: What is the significance of confidence intervals in statistical analysis?

A confidence interval (CI) is like saying:

>>‚ÄúI‚Äôm not 100% sure of the exact number, but I‚Äôm pretty confident it‚Äôs somewhere between this and that.‚Äù

If you estimate the average height of a population from a sample, a confidence interval gives you a range of plausible values for that average ‚Äî along with how confident you are that the true value is inside that range.

>>More formal definition

A confidence interval is an estimated range of values for a population parameter, calculated from sample data, that is likely to contain the true parameter with a specified probability.

For example, a 95% confidence interval means:
If we repeated the sampling process many times, about 95% of those calculated intervals would contain the true parameter.

>>Why it‚Äôs significant

Quantifies uncertainty ‚Äì Instead of a single point estimate, you get a range, making it clear that there‚Äôs natural variability in sampling.

Better decision-making ‚Äì Narrow intervals suggest precise estimates; wide intervals suggest more uncertainty.

Common in reporting ‚Äì Used in research, business analytics, medicine, polling, and more.

Not the same as probability of being correct ‚Äì The ‚Äú95%‚Äù refers to the method‚Äôs long-run success rate, not the probability that your specific interval is correct.

# Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
mean = 50
std_dev = 5
n_samples = 1000

# Generate random numbers from a normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=n_samples)

# Compute mean and standard deviation
calculated_mean = np.mean(data)
calculated_std = np.std(data)

print(f"Calculated Mean: {calculated_mean:.2f}")
print(f"Calculated Standard Deviation: {calculated_std:.2f}")

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Normally Distributed Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(calculated_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {calculated_mean:.2f}')
plt.legend()
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to NumPy array
sales_array = np.array(daily_sales)

# Compute mean and standard deviation
mean_sales = np.mean(sales_array)
std_dev_sales = np.std(sales_array, ddof=1)  # sample std dev (ddof=1)

# Standard error
n = len(sales_array)
standard_error = std_dev_sales / np.sqrt(n)

# z-score for 95% confidence
z_score = stats.norm.ppf(0.975)  # two-tailed, 97.5% point

# Confidence interval
margin_of_error = z_score * standard_error
ci_lower = mean_sales - margin_of_error
ci_upper = mean_sales + margin_of_error

# Output results
print(f"Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```
